# Energy and frequency

With Planck's constant $h$, we write the relation between energy and frequency as:

$$E = hf$$

$$E = hc/\lambda$$

$$ F = G \frac{m_1m_2}{r^2} $$

$$ sin(2x) = 2sin(x)cos(x) = \frac{2tan(x)}{1 + tan^2(x)}$$

$$\displaystyle\sum_{i=0}^{\infty}\frac{1}{4k^2-1}=\frac{1}{2}$$

$$\int_0^\pi\mathrm{e}^{ix}\,\mathrm{d}x=2i$$

```python
from math import atan2, degrees

# ask user for input
x=float(input("Enter value for x:"))
y=float(input("Enter value for y:"))

# calculate angle
angle_rad = atan2(x,y)
angle_deg = degrees(angle_rad)
print("angle with x-axis:", angle_deg, "degrees")

# determine quadrant
if x>0 and y>0:
    print("1st quadrant")
elif x<0 and y>0:
    print ("2nd quadrant")
elif x<0 and y<0:
    print ("3rd quadrant")
elif x>0 and y<0:
    print ("4th quadrant")
```
