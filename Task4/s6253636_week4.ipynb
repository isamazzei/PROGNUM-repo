```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

lengths = np.random.normal(185, 8, 100000)
plt.hist(lengths, bins=40)
plt.show()

sample_mean = np.mean(lengths)
sample_std = np.std(lengths)

print(f"Sample mean: {sample_mean:.2f} cm")
print(f"Sample std: {sample_std:.2f} cm")


plt.figure(figsize=(10,6))
counts, bins, patches = plt.hist(lengths, bins=40, color='skyblue', edgecolor='black', alpha=0.7,label=f"Mean={sample_mean:.2f}, Std={sample_std:.2f}")

plt.xlabel("Height (cm)")
plt.ylabel("Number of People")
plt.title("Distribution of Human Heights")

plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label="Sample mean")
plt.axvspan(sample_mean - sample_std, sample_mean + sample_std, color='yellow', alpha=0.2, label="1 Std Dev Range")

plt.legend()
plt.show()

h = 190    #random height 
prob_below_h = norm.cdf(h, loc=185, scale=8)
print(f"Probability that height ≤ {h} cm: {prob_below_h:.4f}")

prob_within_1std = norm.cdf(185 + 8, loc=185, scale=8) - norm.cdf(185 - 8, loc=185, scale=8)
print(f"Probability of being within 1 std dev: {prob_within_1std:.4f}")
```

```python
import numpy as np 
from matplotlib import pyplot as plt

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

angle=-30
theta=np.radians(angle)

x_rot = x * np.cos(theta) - y * np.sin(theta)
y_rot = x * np.sin(theta) + y * np.cos(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y, color='blue', label='Original square')        
plt.plot(x_rot, y_rot, color='red', label=f'Rotated {angle}°') 
plt.axis('equal') 
plt.grid()
plt.legend()
plt.title(f"Square rotated by {angle}°")
plt.show()
```

```python
import numpy as np
def funcfilter(func,seq):
    arr = np.asarray(seq)
    filtered = arr[arr > 0]
    return func(filtered)

sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)

num = 100
result2 = funcfilter(np.log10, num)
print("Result log10 (single number):", result2)
```

```python
import numpy as np
from astropy.cosmology import Planck18
from scipy import integrate

H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0

Mpc_in_Km= 3.086e19
H0_SI = H0 * 1000 / (3.086e22)

def H(z):
    y1=H0_SI * np.sqrt(Om0*(1+z)**3 + Ode0)
    return y1

def t(zp):
    y2=1/((1+zp)*H(zp))
    return y2

z=2
t_secs= integrate.quad(t,0,z) [0]

secs_Gyr=3.156e16
t_Gyr=t_secs/secs_Gyr

print(f"lookback time at z={z}: {t_Gyr:.2f} Gyr")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

X = np.random.normal(0, 1, 100000)
Y = np.random.normal(0, 1, 100000)


counts, xedges, yedges, im= plt.hist2d(X,Y, bins=30, cmap='Blues', range=[[-5,5],[-5,5]])
plt.colorbar(label='Counts')

xcenters = 0.5 * (xedges[1:] + xedges[:-1])
ycenters = 0.5 * (yedges[1:] + yedges[:-1])

Xg, Yg=np.meshgrid(xcenters, ycenters)

Z= counts/np.sum(counts)

Z_flat= Z.flatten()
Z_sorted=np.sort(Z_flat)[::-1]

cumulative= np.cumsum(Z_sorted)

level_1sigma= Z_sorted[np.searchsorted(cumulative, 0.68)]
level_2sigma= Z_sorted[np.searchsorted(cumulative, 0.95)]
level_3sigma= Z_sorted[np.searchsorted(cumulative, 0.997)]

levels= np.sort ([level_3sigma, level_2sigma, level_1sigma])

contours= plt.contour(Xg, Yg, Z, levels=levels, colors='red', linewidths=2)

fmt= {level_1sigma: "1σ", level_2sigma: "2σ", level_3sigma: "3σ"}

plt.clabel(contours, fmt=fmt, fontsize=11)
plt.grid(True, linestyle='--' , linewidth=1.2, alpha=0.8)

plt.xlabel("x")
plt.ylabel("y")
plt.title("2D histogram + contour")

plt.show                                           
```
