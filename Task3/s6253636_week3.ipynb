```python
import numpy as np
import matplotlib.pyplot as plt

x_range = list(range(-5, 6))
y_range = np.cosh(x_range)

plt.figure()
plt.plot(x_range, y_range, marker='o', linestyle='-', label="y = cosh(x) using range()")

plt.title("Catenary Curve (Hyperbolic Cosine)", fontsize=16)
plt.xlabel("x values", fontsize=14)
plt.ylabel("y = cosh(x)", fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid()
plt.legend(fontsize=12)
plt.show()


x_arange = np.arange(-5, 6, 1)
y_arange = np.cosh(x_arange)
print("Are x values identical?")
print(np.array_equal(np.array(x_range), x_arange))

plt.figure()
plt.plot(x_arange, y_arange, marker='s', linestyle='--', label="y = cosh(x) using arange()")

plt.title("Catenary Curve (NumPy arange)", fontsize=16)
plt.xlabel("x values", fontsize=14)
plt.ylabel("y = cosh(x)", fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid()
plt.legend(fontsize=12)

plt.show()
```
